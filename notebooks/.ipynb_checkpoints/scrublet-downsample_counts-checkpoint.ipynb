```python
data_dir = 'sce_normalized_data_inflate'
import scipy.sparse as scs
from os import listdir
from os.path import isfile, join
import os
from plot_results import get_dbl_metrics
save=True
from scipy.io import mmread
```

```python
import sys
import numpy as np
from numpy import *
import pandas as pd

from scipy import sparse
from sklearn import preprocessing
import pathlib as pl

import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.linear_model import Lasso

import scrublet as scr
from sklearn.metrics import precision_recall_curve, accuracy_score, average_precision_score


##########
#functions


##########
#READ IN DATA
#file = pl.Path(sys.argv[1])
#dat = np.load(file)
#ano_path = sys.argv[2]
#ano = pd.read_csv(ano_path)
#y = np.squeeze(np.array((ano.doublet_true_labels=='Doublet')|(ano.doublet_true_labels=='doublet')).astype(float))
#data_name = sys.argv[3]
#save_path = sys.argv[4] + '/scrublet_out/' + data_name

fracs = [0.95, 0.9, 0.75, 0.5, 0.25, 0.1, 0.05]
#fracs = [0.95, 0.9]

files = ['HEK-HMEC-MULTI_real_logcounts.mtx',
       'HMEC-orig-MULTI_real_logcounts.mtx',
       'HMEC-rep-MULTI_real_logcounts.mtx', 'J293t-dm_real_logcounts.mtx',
       'cline-ch_real_logcounts.mtx', 'hm-12k_real_logcounts.mtx',
       'hm-6k_real_logcounts.mtx', 'mkidney-ch_real_logcounts.mtx',
       'nuc-MULTI_real_logcounts.mtx', 'pbmc-1A-dm_real_logcounts.mtx',
       'pbmc-1B-dm_real_logcounts.mtx', 'pbmc-1C-dm_real_logcounts.mtx',
       'pbmc-2ctrl-dm_real_logcounts.mtx',
       'pbmc-2stim-dm_real_logcounts.mtx', 'pbmc-ch_real_logcounts.mtx',
       'pdx-MULTI_real_logcounts.mtx']
files = ['hm-12k_real_logcounts.mtx',
       'hm-6k_real_logcounts.mtx', 'mkidney-ch_real_logcounts.mtx',
       'nuc-MULTI_real_logcounts.mtx', 'pbmc-1A-dm_real_logcounts.mtx',
       'pbmc-1B-dm_real_logcounts.mtx', 'pbmc-1C-dm_real_logcounts.mtx',
       'pbmc-2ctrl-dm_real_logcounts.mtx',
       'pbmc-2stim-dm_real_logcounts.mtx', 'pbmc-ch_real_logcounts.mtx',
       'pdx-MULTI_real_logcounts.mtx']
reps = range(1,10)
#reps = [0]

#reps = range(1)
#files = ['J293t-dm_real_logcounts.mtx']
#fracs = [0.05]



for file in files:

    data_name = file[:-19]
    print(data_name) 

    ano_path  = '../data/'+ data_dir + '/' + data_name + '_anno.csv'

    #- READ IN BARCODE ANNOTATIONS
    ano = pd.read_csv(ano_path)
    true = pd.factorize(ano.x)[0]
    labels = ano.x
    if (labels[0]=='doublet'):
        tmp = true + 3
        tmp[tmp==3] = 1
        tmp[tmp==4] = 0
        true = tmp

    for frac in fracs:
        for i in reps:
            
            mtx_path = '../results_PU/downsample_counts/vaeda/' + data_name + '_frac' + str(frac) + '_rep' + str(i) + '_X.mtx'
            
            Xr = mmread(mtx_path)
            Xr = scs.csr_matrix(Xr).toarray()

            save_path = '../results_PU/downsample_counts/scrublet'
            if not os.path.exists(save_path):
                os.makedirs(save_path)
            save_path = save_path + '/' + data_name + '_frac' + str(frac) + '_rep' + str(i) + '_'
            print(save_path)


            ##########
            #SCRUBLET
            scrub = scr.Scrublet(Xr)
            doublet_scores, predicted_doublets = scrub.scrub_doublets()
            
            if np.any(predicted_doublets==None):
                res = pd.DataFrame(np.vstack([true, doublet_scores]).T, columns=['ano', 'doublet_scores'])
            else:
                res = pd.DataFrame(np.vstack([true, doublet_scores, predicted_doublets]).T, columns=['ano', 'doublet_scores', 'calls'])
            res.to_csv(save_path + 'scores.csv') 
            
            
            roc, apauc, ap = get_dbl_metrics(true, doublet_scores)
            results = pd.DataFrame(data=[[roc,apauc,ap]], columns=['AUROC','AUPRC','AP'])

            print(results)

            #SAVE SCORES
            tmp1 = np.zeros((len(labels), 2))
            df = pd.DataFrame(tmp1, index=labels, columns=['annotation', 'score'])
            df.annotation = labels
            df.score = doublet_scores
            #if(save):
            #    df.to_csv(save_path + data_name + '_scores.csv') 

            #PR and ROC curves
            plt.figure(4)
            res = get_dbl_metrics(true, doublet_scores)
            plt.show()
            plt.close()

            #save AUCs
            hm_pr = pd.DataFrame(np.array(res).T, index=['AUROC', 'AUPRC', 'AP']).T
            if(save):
                hm_pr.to_csv(save_path + 'scores_ROC_PR_area_ALL.csv') 


```
