```python

def sim_coin_flip(X, frac_doublets=None):
    
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    res = X1 + X2 

    flips = np.random.randint(2, size=(X1.shape[0]))
    
    lib_sze = np.sum(X1[flips == 0,:], axis=1)
    ls = np.sum(res[flips == 0,:], axis=1)
    sf = lib_sze / ls
    res[flips == 0,:] = np.multiply(res[flips == 0,:].T, sf).T
    
    lib_sze = np.sum(X2[flips == 1,:], axis=1)
    ls = np.sum(res[flips == 1,:], axis=1)
    sf = lib_sze / ls
    res[flips == 1,:] = np.multiply(res[flips == 1,:].T, sf).T
    
    ind = np.arange(X1.shape[0])
    np.random.shuffle(ind)
    res = res[ind,:]
    
    return res[:num_doublets,:]


def sim_biggest(X, frac_doublets=None):
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    res = X1 + X2 

    lib1 = np.sum(X1, axis=1)
    lib2 = np.sum(X2, axis=1)

    lib_sze = np.vstack([lib1, lib2])
    biggest = np.argmax(lib_sze, axis=0)

    ls = np.sum(res[biggest==0,:], axis=1)
    sf = lib1[biggest==0] / ls
    res[biggest == 0,:] = np.multiply(res[biggest == 0,:].T, sf).T

    ls = np.sum(res[biggest==1,:], axis=1)
    sf = lib2[biggest==1] / ls
    res[biggest == 1,:] = np.multiply(res[biggest == 1,:].T, sf).T
    
    ind = np.arange(X1.shape[0])
    np.random.shuffle(ind)
    res = res[ind,:]
    
    return res[:num_doublets,:]

def sim_trips(X, frac_doublets=None):
    
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    X3 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    np.random.shuffle(X3)
    
    res = X1 + X2 + X3
        
    lib1 = np.sum(X1, axis=1)
    lib2 = np.sum(X2, axis=1)
    lib3 = np.sum(X3, axis=1)

    lib_sze = np.vstack([lib1, lib2, lib3])
    biggest = np.argmax(lib_sze, axis=0)

    ls = np.sum(res[biggest==0,:], axis=1)
    sf = lib1[biggest==0] / ls
    res[biggest == 0,:] = np.multiply(res[biggest == 0,:].T, sf).T

    ls = np.sum(res[biggest==1,:], axis=1)
    sf = lib2[biggest==1] / ls
    res[biggest == 1,:] = np.multiply(res[biggest == 1,:].T, sf).T

    ls = np.sum(res[biggest==2,:], axis=1)
    sf = lib3[biggest==2] / ls
    res[biggest == 2,:] = np.multiply(res[biggest == 2,:].T, sf).T
    
    return res[:num_doublets,:]


def sim_sum(X, frac_doublets=None):
    
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    res = X1 + X2 

    ind = np.arange(X1.shape[0])
    np.random.shuffle(ind)
    res = res[ind,:]
    
    return res[:num_doublets,:]

def sim_avg(X, frac_doublets=None):
    
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    res = (X1 + X2) / 2

    ind = np.arange(X1.shape[0])
    np.random.shuffle(ind)
    res = res[ind,:]
    
    return res[:num_doublets,:]

def sim_inflate(X, frac_doublets=None):
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    res = X1 + X2 
    
    lib1 = np.sum(X1, axis=1)
    lib2 = np.sum(X2, axis=1)
    
    lib_sze = np.maximum.reduce([lib1,lib2])
    high = np.max(lib_sze)
    
    inflated_sze = np.zeros([len(lib_sze)])
    for i, low in enumerate(lib_sze): 
        inflated_sze[i] = random.choice(lib_sze[lib_sze>=low])
        #inflated_sze[i] = np.random.randint(low, high=high+1, dtype=int)
    
    ls = np.sum(res, axis=1)
    sf = inflated_sze / ls
    res= np.multiply(res.T, sf).T
    
    return res[:num_doublets,:]


def sim_lib_ard_var(X, p=0.3, frac_doublets=None):
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    res = X1 + X2 
    
    lib1 = np.sum(X1, axis=1)
    lib2 = np.sum(X2, axis=1)
    
    lib_sze = np.maximum.reduce([lib1,lib2])
    high = np.max(lib_sze)
    
    inflated_sze = np.zeros([len(lib_sze)])
    for i, low in enumerate(lib_sze): 
        inflated_sze[i] = random.choice(lib_sze[lib_sze>=low])
    
    ls = np.sum(res, axis=1)
    sf = inflated_sze / ls
    res= np.multiply(res.T, sf).T
    
    zeros = np.logical_or(np.logical_and(X1==0, X2!=0), np.logical_and(X1!=0, X2==0))
    cnt = 0
    for i in range(res.shape[0]):
        for j in range(res.shape[1]):    
            if(zeros[i,j]):
                rn = np.random.random() # 1 in 100 chance of zeroing out
                if (rn < p):
                    cnt += 1
                    res[i,j] = 0
    print(cnt)
    np.random.shuffle(res)
    
    return res[:num_doublets,:]

def sim_multi(X, frac_doublets=None):
    
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    X_t = X1 + X2 
    
    res = np.zeros(X_t.shape)
    for i in range(X_t.shape[0]):
        zero_mask = X_t[i,:] != 0 #just taking out the zeros to save some time
        prob = X_t[i,zero_mask]
        prob /= np.sum(prob) #converting to probability
        dd = int((sum(X_t[i,:]))) / 2 #library size
        sim_counts =  multinomial.rvs(n=dd, p=prob)
        res[i,zero_mask] = sim_counts 
    
    np.random.shuffle(res)
        
    return res[:num_doublets,:]

def sim_inflate_multi(X, frac_doublets=None):
    
    if frac_doublets==None:
        num_doublets = 1 * X.shape[0]
    else:
        num_doublets = int(frac_doublets * X.shape[0])
    
    X1 = np.copy(X)
    X2 = np.copy(X)
    np.random.shuffle(X1)
    np.random.shuffle(X2)
    
    X_t = X1 + X2 
    
    lib1 = np.sum(X1, axis=1)
    lib2 = np.sum(X2, axis=1)
    
    #lib_sze = np.maximum.reduce([lib1,lib2])
    lib_sze = np.sum(X_t, axis = 1) / 2
    
    res = np.zeros(X_t.shape)
    for i in range(X_t.shape[0]):
        zero_mask = X_t[i,:] != 0 #just taking out the zeros to save some time
        prob = X_t[i,zero_mask]
        prob /= np.sum(prob) #converting to probability
        lib = int(random.choice(lib_sze[lib_sze>=lib_sze[i]]))  #library size
        sim_counts =  multinomial.rvs(n=lib, p=prob)
        res[i,zero_mask] = sim_counts 
    
    np.random.shuffle(res)
        
    return res[:num_doublets,:]
```
